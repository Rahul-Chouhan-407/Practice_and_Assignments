```python
import pandas as pd

# Create a Series
data = pd.Series([1, 2, 3, 4, 5])

# Define a custom function
def square(x):
    return x ** 2

# Apply the function to each element in the Series
result = data.apply(square)
print(result)
```

```python
import pandas as pd

# Create a Series with int64 dtype
data = pd.Series([1, 2, 3, 4, 5], dtype='int64')

# Convert the Series to int32 dtype
data = data.astype('int32')

# Check the data type of the Series
print(data.dtype)  # Output: int32
```

```python
# Create a DataFrame
data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(data)

# Define a custom function
def multiply_by_two(column):
    return column * 2

# Apply the function to each column in the DataFrame
result = data.apply(multiply_by_two)
print(result)
```

```python
import pandas as pd

# Create a DataFrame
data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Define a custom function
def sum_of_row(row):
    return row['A'] + row['B']

# Apply the function to each row in the DataFrame
result = data.apply(sum_of_row, axis=1)
print(result)
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/Hotel Bookings.csv", usecols=['hotel', 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'lead_time',	'stays_in_weekend_nights',	'stays_in_week_nights',	'adults',	'children',	'babies', 'country', 'customer_type', 'reservation_status', 'reservation_status_date'], converters = {'arrival_date_month': month_name_to_int}, parse_dates = ["reservation_status_date"])
```

```python
def month_name_to_int(month_name):
  month_mapping = {'January': 1,'February': 2,'March': 3,'April': 4,'May': 5,'June': 6,'July': 7,'August': 8,'September': 9,'October': 10,'November': 11,'December': 12}
  return month_mapping.get(month_name, month_name)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.models import Sequential
from tensorflow.keras.models import Sequential, Model, load_model, save_model
from tensorflow.keras.layers import Dense, Activation, Dropout, Input, Masking, TimeDistributed, LSTM, Conv1D, Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import optimizers
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Assignments/Neural Network Assignment - Intellipaat/Worksheet in Assignment - Titanic.xlsx")
```

```python
df.head(100)
```

```python
# Remove unnecessary columns
df = df.drop(['name', 'ticket', 'cabin', 'embarked', 'boat', 'body', 'home_dest'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Fill missing values with the mean
df = df.fillna(df.mean())
```

```python
# Convert categorical variables to numerical variables
labelencoder = LabelEncoder()
df['sex'] = labelencoder.fit_transform(df['sex'])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(['survived'], axis=1), df['survived'], test_size=0.2, random_state=42)
```

```python
# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define the neural network architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test accuracy:', test_acc)
```

**Database used:** diabetes.csv

**Tasks to be done:**

A. Build a sequential model using Keras on top of this Diabetes dataset to find out if the patient has diabetes or not, using ‘Pregnancies’, ‘Glucose’ & ‘BloodPressure’ as independent columns.

a. This model should have 1 hidden layer with 8 nodes

b. Use Stochastic Gradient as the optimization algorithm

c. Fit the model, with number of epochs to be 100 and batch size to be 10

B. Build another sequential model where ‘Outcome’ is the dependent variable and all other columns are predictors.

a. This model should have 3 hidden layers with 16 nodes in each layer

b. Use ‘adam’ as the optimization algorithm

c. Fit the model, with number of epochs to be 150 and batch size to be 10

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load the diabetes dataset
diabetes_df = pd.read_csv('diabetes.csv')
```

**First Model**

```python
# Select the features for the first model
features1 = ['Pregnancies', 'Glucose', 'BloodPressure', 'Outcome']
diabetes1_df = diabetes_df[features1].copy()
```

```python
# Split the data into training and testing sets
X_train1, X_test1, y_train1, y_test1 = train_test_split(diabetes1_df.drop('Outcome', axis=1), diabetes1_df['Outcome'], test_size=0.2, random_state=42)
```

```python
# Standardize the data
scaler1 = StandardScaler()
X_train1 = scaler1.fit_transform(X_train1)
X_test1 = scaler1.transform(X_test1)
```

```python
# Build the first model
model1 = keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=(3,)),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model1.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model1.fit(X_train1, y_train1, epochs=100, batch_size=10, verbose=1)
```

```python
# Evaluate the model
score1 = model1.evaluate(X_test1, y_test1, verbose=0)
print('Test loss:', score1[0])
print('Test accuracy:', score1[1])
```

**Second Model**

```python
# Select the features for the second model
features2 = list(diabetes_df.columns)
features2.remove('Outcome')
diabetes2_df = diabetes_df[features2].copy()
```

```python
# Split the data into training and testing sets
X_train2, X_test2, y_train2, y_test2 = train_test_split(diabetes2_df, diabetes_df['Outcome'], test_size=0.2, random_state=42)
```

```python
# Standardize the data
scaler2 = StandardScaler()
X_train2 = scaler2.fit_transform(X_train2)
X_test2 = scaler2.transform(X_test2)
```

```python
# Build the second model
model2 = keras.Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=(8,)),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model2.fit(X_train2, y_train2, epochs=150, batch_size=10, verbose=1)
```

```python
# Evaluate the model
score2 = model2.evaluate(X_test2, y_test2, verbose=0)
print('Test loss:', score2[0])
print('Test accuracy:', score2[1])
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the wine dataset
wine_data = load_wine()
```

```python
# Create a pandas dataframe with the data and feature names as column headers
df_wine = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
```

```python
# Add target variable to the dataframe
df_wine['target'] = wine_data.target
```

**Checking out the Data**

```python
# Check out the data
df_wine.head()
```

Split the data into Training and Testing Datasets

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_wine.iloc[:,:-1], df_wine['target'], test_size=0.2, random_state=42)
```

**Scaling the Data**

```python
# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Training the Model**

```python
# Train your model
model = MLPClassifier(hidden_layer_sizes=(64,32), max_iter=50)
model.fit(X_train, y_train)
```

**Using predict () method to get predictions**

```python
# Use predict() method to get predictions
y_pred = model.predict(X_test)
```

**Evaluating the Model**

```python
# Use SciKit-Learn's built-in metrics to evaluate how well our model performed
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

**Extracting the MLP weights and biases after training your model**

```python
# Extract the MLP weights and biases after training your model
weights = model.coefs_
biases = model.intercepts_
print("Weights:", weights)
print("Biases:", biases)
```

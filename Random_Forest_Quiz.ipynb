```python
# Filter the population with income greater than 50%
greater_than_50k = df[df['Unnamed: 14'] == '>50K']

# Get the total count of population with income greater than 50%
total_population = len(greater_than_50k)

print('Total population with income greater than 50% income:', total_population)
```

```python
# select samples that are unmarried and work less than 20 hours per week
samples = df[(df['marital-status'] == 'Never-married') & (df['hours-per-week'] < 20)]

# compute the number of samples
num_samples = len(samples)

# print the result
print(f"Number of samples: {num_samples}")
```

```python
# Group the data by country and get the total population for each country
populations = df.groupby('native-country')['population'].sum()

# Find the country with the highest population
highest_population_country = populations.idxmax()

# Find the country with the lowest population
lowest_population_country = populations.idxmin()

print("Country with the highest population:", highest_population_country)
print("Country with the lowest population:", lowest_population_country)
```
